```julia

x = slider(0:.1:2pi, label="x")
s = slider(-1:.05:1, value=map(a->sin(2a), observe(x)), label="sin(2x)")
c = slider(-1:.05:1, value=map(a->cos(2a), observe(x)), label="cos(2x)")
map(display, [x,s,c]);
```
